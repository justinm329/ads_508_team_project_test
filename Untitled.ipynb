```python
orders_df = pd.read_csv('Data/orders_dataset.csv')
orders_df.head()
```

```python
products_df = pd.read_csv('Data/products_dataset.csv')
products_df.head()
products_df.isnull().sum()
```

```python
location_df = pd.read_csv('Data/geolocation_dataset.csv')
location_df.head()
location_df.isnull().sum()
```

```python
translate_df = pd.read_csv('Data/product_category_name_translation.csv')
translate_df.head()
```

```python
## slice out columns needed
#sliced_customers = customer_df[['customer_zip_code_prefix', 'customer_city', 'customer_state']]
sliced_orders = orders_df[['order_id', 'customer_id', 'order_status', "order_purchase_timestamp"]]
sliced_items = order_items_df[['order_id', 'price', 'freight_value', 'product_id']]
sliced_payments = order_payments_df[['order_id', 'payment_type', 'payment_installments', 'payment_value']]
sliced_reviews = order_reviews_df[['order_id', 'review_id' , 'review_score']]
sliced_products = products_df[['product_id','product_category_name']]
sliced_locations = location_df[['geolocation_zip_code_prefix', 'geolocation_city', 'geolocation_state']]
sliced_translate = translate_df[['product_category_name', 'product_category_name_english']]
```

```python
sliced_products.describe()
sliced_translate.
```

```python
# perform merge or inner join on customer_df
df_merged = customer_df.merge(
    sliced_orders, on = 'customer_id').merge(
    sliced_items, on = 'order_id').merge(
    sliced_payments, on = 'order_id').merge(
    sliced_reviews, on = 'order_id').merge(
    sliced_products, on = 'product_id').merge(
    sliced_translate, on = 'product_category_name')
df_merged.head()
```

```python
corr_matrix = df_merged[["product_category_name_english", "freight_value", "order_status", "payment_value", "payment_type", "review_score"]].corr()

# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix for Merged DataFrame")
plt.show()
```

```python
# Create a boxplot using Seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(data=order_reviews_df, y='review_score')
plt.title('Boxplot of Review Scores Distribution')
plt.xlabel('File')
plt.ylabel('Review Score Values')
plt.grid(True)
plt.show()
```
